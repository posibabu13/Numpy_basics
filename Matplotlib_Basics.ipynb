```python
import numpy as np
t=np.arange(0,5,0.2)
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**4,'g^',label='^4')
plt.grid()
plt.legend()
plt.show()
```

```python
x1=[1,2,4,5]
y1=[1,4,16,25]
x2=[1,2,3,4]
y2=[1,4,6,8]

lines=plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='r',linewidth=10.0)
plt.setp(lines[1],color='g',linewidth=5.0)
plt.grid()
plt.show()
```

```python
def f(t):
    return np.exp(-t**2) * np.cos(2*np.pi*t)


t1=np.arange(1,5,0.1)
t2=np.arange(2,8,0.5)

plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t2,f(t2),'b--',linewidth =5)


plt.subplot(212)
plt.grid()
plt.plot(t1,f(t1),'go',linewidth=10)
plt.show()
```
